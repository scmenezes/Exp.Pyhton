```python
financials_dict = {}
for ativo in ativoslist:
    info = yf.Ticker(ativo)
    financials_dict[ativo] = pd.DataFrame(info.financials)
```

```python
#Desafio- Parte 1. Armazenar em uma estrutura de dados de sua escolha
```

```python
financials_last_years = {}
for ativo in financials_dict.keys():
    if '2023-12-31' in financials_dict[ativo].columns:
        print('ok')
        df_aux = pd.DataFrame(financials_dict[ativo]['2023-12-31'])

    else:
        print(ativo)
        print(financials_dict[ativo].columns[0])
        df_aux = pd.DataFrame(financials_dict[ativo]['2023-09-30'])

    financials_last_years[ativo] = df_aux
```

```python
for ativo in financials_last_years.keys():

    if '2023-12-31' in financials_last_years[ativo].columns:
        financials_last_years[ativo][ativo] = financials_last_years[ativo]['2023-12-31']
        financials_last_years[ativo].drop(financials_last_years[ativo].columns[0], axis = 1, inplace=True)
    else:
        financials_last_years[ativo][ativo] = financials_last_years[ativo]['2023-09-30']
        financials_last_years[ativo].drop(financials_last_years[ativo].columns[0], axis = 1, inplace=True)
```

```python
lista = []
for ativo in financials_last_years.keys():
    lista.append(financials_last_years[ativo])
```

```python
lista
```

```python
df = pd.concat(lista, axis = 1)

df
```
